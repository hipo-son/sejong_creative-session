```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.constants import g
# 초깃값
x,y=0,0
v0=10
theta=np.pi/4
dt=1e-3
vx,vy=v0*np.cos(theta),v0*np.sin(theta)

x_point,y_point=[x],[y]
C=1
while C:
    vx=vx+0*dt
    vy=vy+(-g)*dt
    x=x+vx*dt
    y=y+vy*dt
    C=(y>0)
    x_point.append(x)
    y_point.append(y)

plt.plot(x_point,y_point)
plt.show()
```

```python
from scipy.constants import g
# 각도를 랜던으로 조금씩 바꾸기
z=0
while z<10:
    # 초깃값
    x,y=0,0
    v0=10
    theta=np.pi/4+np.random.rand()*1e-1
    dt=1e-3
    vx,vy=v0*np.cos(theta),v0*np.sin(theta)

    x_point,y_point=[x],[y]
    C=1
    while C:
        vx=vx+0*dt
        vy=vy+(-g)*dt
        x=x+vx*dt
        y=y+vy*dt
        C=(y>0)
        x_point.append(x)
        y_point.append(y)

    plt.plot(x_point,y_point)
    z+=1

plt.show()
```
